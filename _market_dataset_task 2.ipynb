```python
# Convert 'Acquisition_Cost' to numeric
df['Acquisition_Cost'] = df['Acquisition_Cost'].replace('[\$,]', '', regex=True).astype(float)
```

```python
# Extract numerical values from 'Duration'
df['Duration'] = df['Duration'].str.extract('(\d+)').astype(float)
```

```python
# Compute Click-Through Rate (CTR) and Cost Per Click (CPC)
df["CTR"] = (df["Clicks"] / df["Impressions"]) * 100  # CTR in percentage
df["CPC"] = df["Acquisition_Cost"] / df["Clicks"]  # CPC in dollars
```

```python
# Summary statistics for new metrics
ctr_cpc_summary = df[["CTR", "CPC"]].describe()
print("CTR & CPC Summary:\n", ctr_cpc_summary)
```

```python
# Aggregate key metrics by marketing channel
channel_performance = df.groupby("Channel_Used").agg(
    Avg_CTR=("CTR", "mean"),
    Avg_CPC=("CPC", "mean"),
    Avg_ROI=("ROI", "mean"),
    Total_Clicks=("Clicks", "sum"),
    Total_Impressions=("Impressions", "sum")
).reset_index()
```

```python
# Sort by highest ROI
channel_performance = channel_performance.sort_values(by="Avg_ROI", ascending=False)
print("Marketing Channel Performance:\n", channel_performance)
```

```python
# Clean the Acquisition_Cost column (remove $ and convert to float)
df['Acquisition_Cost'] = df['Acquisition_Cost'].replace('[\$,]', '', regex=True).astype(float)

# Calculate CTR (Click-Through Rate) and CPC (Cost Per Click)
df['CTR'] = (df['Clicks'] / df['Impressions']) * 100  # CTR in percentage
df['CPC'] = df['Acquisition_Cost'] / df['Clicks']  # Cost Per Click

# Display the first few rows to confirm the changes
df.head()
```

```python
# Re-run code snippet to generate the visualizations due to reset

import matplotlib.pyplot as plt
import seaborn as sns

# Visualizations for the user to download
# Visualization 1: Bar Chart - CTR, CPC, and Conversion Rate by Marketing Channel
channel_performance = df.groupby('Channel_Used').agg({
    'CTR': 'mean',
    'CPC': 'mean',
    'Conversion_Rate': 'mean'
}).reset_index()

fig, ax = plt.subplots(figsize=(10, 6))
channel_performance.set_index("Channel_Used")[["CTR", "CPC", "Conversion_Rate"]].plot(kind="bar", ax=ax)
plt.title("CTR, CPC, and Conversion Rate by Marketing Channel")
plt.ylabel("Percentage / Cost ($)")
plt.xlabel("Marketing Channel")
plt.xticks(rotation=45)
plt.legend(["CTR (%)", "CPC ($)", "Conversion Rate (%)"])
bar_chart_path = "/mnt/data/Bar_Chart_CTR_CPC_Conversion_Rate.png"
# fig.savefig(bar_chart_path, bbox_inches="tight", dpi=300)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

# Check if the conversion worked
df['Date'].dtype

# Recreate Visualization 2: Line Graph - ROI Performance Over Time
df["YearMonth"] = df["Date"].dt.to_period("M")
roi_trend = df.groupby("YearMonth")["ROI"].mean().reset_index()

fig, ax = plt.subplots(figsize=(10, 5))
plt.plot(roi_trend["YearMonth"].astype(str), roi_trend["ROI"], marker="o", linestyle="-", color="b")
plt.title("ROI Performance Over Time")
plt.xlabel("Month-Year")
plt.ylabel("Average ROI")
plt.xticks(rotation=45)
line_graph_path = "/mnt/data/Line_Graph_ROI_Performance_Over_Time.png"
# fig.savefig(line_graph_path, bbox_inches="tight", dpi=300)
```

```python
# Visualization 3: Heatmap - Conversion Rate by Location
location_performance = df.groupby('Location').agg({
    'Conversion_Rate': 'mean'
}).reset_index()

fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(location_performance.set_index('Location')[['Conversion_Rate']], 
            annot=True, cmap='magma', linewidths=0.5, fmt=".2%")
plt.title("Conversion Rate by Location (Heatmap)")
plt.ylabel("Location")
plt.xlabel("Conversion Rate")
heatmap_path = "/mnt/data/Heatmap_Conversion_Rate_by_Location.png"
# fig.savefig(heatmap_path, bbox_inches="tight", dpi=300)
```

```python
# Visualization 4: Scatter Plot - ROI vs. Conversion Rate
fig, ax = plt.subplots(figsize=(8, 5))
sns.scatterplot(data=df, x="Conversion_Rate", y="ROI", alpha=0.5)
plt.title("ROI vs. Conversion Rate")
plt.xlabel("Conversion Rate (%)")
plt.ylabel("Return on Investment (ROI)")
# scatter_plot_path = "/mnt/data/Scatter_Plot_ROI_vs_Conversion_Rate.png"
# fig.savefig(scatter_plot_path, bbox_inches="tight", dpi=300)
```

```python
# Visualization 5: Density Plot - Distribution of ROI
fig, ax = plt.subplots(figsize=(8, 5))
sns.kdeplot(df['ROI'], shade=True, color="blue")
plt.title("Density Plot: ROI Distribution")
plt.xlabel("ROI")
plt.ylabel("Density")
density_plot_path = "/mnt/data/Density_Plot_ROI_Distribution.png"
# fig.savefig(density_plot_path, bbox_inches="tight", dpi=300)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Calculate the average metrics for each channel
channel_performance = df.groupby("Channel_Used")[["ROI", "CPC", "CTR"]].mean().reset_index()
channel_performance.rename(columns={'ROI': 'Avg_ROI', 'CPC': 'Avg_CPC', 'CTR': 'Avg_CTR'}, inplace=True)

# Set modern, visually appealing style
sns.set_style("darkgrid")
plt.figure(figsize=(18, 10))

# Create a 2x2 subplot layout
fig, axes = plt.subplots(2, 2, figsize=(18, 12))
fig.suptitle("Marketing Channel Performance - Visuals", fontsize=18, fontweight='bold', color="darkblue")

# --- 1. Bar Chart for ROI --- #
sns.barplot(x="Avg_ROI", y="Channel_Used", data=channel_performance, ax=axes[0, 0], palette="Blues_d")
axes[0, 0].set_title("ðŸ“ˆ Average ROI by Channel", fontsize=14, fontweight="bold")
axes[0, 0].set_xlabel("ROI", fontsize=12)
axes[0, 0].set_ylabel("Channel", fontsize=12)

# --- 2. Scatter Plot for CTR vs CPC --- #
scatter = sns.scatterplot(
    x="Avg_CPC", y="Avg_CTR", data=channel_performance, hue="Channel_Used", palette="plasma", s=200, ax=axes[0, 1]
)
axes[0, 1].set_title("ðŸŽ¯ CTR vs CPC by Channel", fontsize=14, fontweight="bold")
axes[0, 1].set_xlabel("Cost Per Click (CPC)", fontsize=12)
axes[0, 1].set_ylabel("Click-Through Rate (CTR)", fontsize=12)
axes[0, 1].legend(title="Channel")

# --- 3. Line Chart for CTR Trend across Channels --- #
sns.lineplot(x="Channel_Used", y="Avg_CTR", data=channel_performance, marker="o", linewidth=3, markersize=10,
             ax=axes[1, 0], color="crimson")
axes[1, 0].set_title("ðŸ“Š CTR Trend Across Channels", fontsize=14, fontweight="bold")
axes[1, 0].set_xlabel("Marketing Channels", fontsize=12)
axes[1, 0].set_ylabel("Average CTR (%)", fontsize=12)
axes[1, 0].tick_params(axis="x", rotation=30)

# --- 4. Heatmap for Key Metrics Correlation --- #
corr_matrix = df[["ROI", "CTR", "CPC", "Clicks", "Impressions"]].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", linewidths=0.5, ax=axes[1, 1])
axes[1, 1].set_title("ðŸ”¥ Correlation Heatmap of Key Metrics", fontsize=14, fontweight="bold")

# Adjust layout for better spacing
plt.tight_layout(rect=[0, 0, 1, 0.96])
# plt.show()
plt.savefig('plot.png')
```
